```
properties = response_json['properties']
```

```
import pandas as pd
CITY_NAME_DF = pd.DataFrame(properties)
```

```
anaheim_listings = CITY_NAME_DF[['property_id', 'prop_type', 'address', 'price', 'baths_half', 'baths_full', 'beds', 'building_size']]
anaheim_listings["baths_half"] = anaheim_listings["baths_half"].fillna(0)
anaheim_listings
```

```
from pandas.io.json import json_normalize
anaheim_address = json_normalize(anaheim_listings['address'])
```

```
anaheim_condo_df = pd.concat([anaheim_listings,anaheim_address],axis=1)
anaheim_condo_df.dropna(subset=['building_size'], inplace=True)
```

```
building_size = json_normalize(anaheim_df['building_size'])
building_size
```

```
anaheim_condo_df = pd.concat([anaheim_listings, anaheim_address, building_size],axis=1)
anaheim_condo_df = pd.DataFrame(anaheim_condo_df)
```

```
anaheim_condo_df.drop(['address', 'building_size', 'state', 'fips_code', 'lat', 'lon', 'neighborhood_name', 'is_approximate', 'time_zone', 'units'], axis=1, inplace=True)
anaheim_condo_df.rename(columns={'size': "building_size"}, inplace=True)
```
